```python
import requests
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(url)
page.text

from bs4 import BeautifulSoup
soup = BeautifulSoup(page.text,'lxml')
```

```python
review_text = []
my_table = soup.find('table',{'class':'wikitable sortable'})
links = my_table.findAll('td')
i = 0
postcode_text = []
borough_text = []
neighbourhood_text = []
for link in links:
    if (i % 3) == 0:  
        postcode_text.append(link.get_text())
    elif (i % 3) == 1:
        borough_text.append(link.get_text())
    else:
        neighbourhood_text.append(link.get_text()[:-1])  
    i+=1
final_array = []
for x, y ,z in zip(postcode_text, borough_text,neighbourhood_text):
    final_array.append({'Postalcode':x,
                        'Borough':y,
                        'Neighbourhood':z
                        })
df = pd.DataFrame(final_array)
df.replace({'Not assigned':np.nan },inplace = True)
df = df.dropna()
df.reset_index(drop=True,inplace = True)
df = df.groupby(['Postalcode','Borough'],as_index = False,sort=False).agg(lambda x: ','.join(x))
df
```
